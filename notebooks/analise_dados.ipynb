### 1. Biblioteca Pandas (`pd`)

#### 1.1 Carregamento e Visualização de Dados
*   `pd.read_excel('/caminho/do/arquivo.xlsx')`: Carrega dados de um arquivo Excel para um DataFrame.
*   `pd.read_csv('/caminho/do/arquivo.csv', encoding='latin1', sep=';')`: Carrega dados de um arquivo CSV, especificando a codificação e o separador.
*   `dados.head()`: Exibe as primeiras 5 linhas do DataFrame.
*   `dados.columns`: Lista os nomes de todas as colunas do DataFrame.
*   `dados.info()`: Fornece um resumo conciso do DataFrame, incluindo tipo de dados e contagem de valores não nulos.
*   `display(df)`: Exibe o DataFrame completo (ou parte dele, dependendo da configuração).
*   `pd.read_sql(query, con=conexao)`: Executa uma consulta SQL e carrega os resultados diretamente para um DataFrame do Pandas.

#### 1.2 Filtragem e Seleção de Dados
*   `dados['COLUNA']`: Seleciona uma única coluna do DataFrame.
*   `dados[dados['COLUNA'] == 'VALOR']`: Filtra linhas onde 'COLUNA' tem um 'VALOR' específico.
*   `dados[(condição1) & (condição2)]`: Filtra linhas com múltiplas condições (use `|` para 'OU').
*   `dados[['COLUNA1', 'COLUNA2']]`: Seleciona múltiplas colunas.

#### 1.3 Agrupamento e Contagem
*   `dados.groupby('COLUNA')['ID'].nunique()`: Agrupa por 'COLUNA' e conta o número de IDs únicos em cada grupo.
*   `dados.groupby('COLUNA', dropna=False)['ID'].nunique()`: Inclui valores nulos no agrupamento para contagem de IDs únicos.
*   `dados['COLUNA'].value_counts(dropna=False)`: Conta a frequência de valores únicos em uma coluna, incluindo nulos.
*   `pd.pivot_table(dados, values=['VALOR'], index =['INDICE'], columns=['COLUNAS'], aggfunc='count')`: Cria uma tabela dinâmica para análise cruzada.

#### 1.4 Tratamento de Valores Faltantes (NaN)
*   `dados['COLUNA'].isnull().sum()`: Conta o número de valores nulos em uma coluna.
*   `dados['COLUNA'].fillna('VALOR_PADRAO')`: Preenche valores nulos em 'COLUNA' com um 'VALOR_PADRAO'.
*   `dados.loc[condição, 'COLUNA'] = NOVO_VALOR`: Atribui 'NOVO_VALOR' a linhas que satisfazem a 'condição' na 'COLUNA' específica.
*   `pd.to_datetime(df, errors='coerce')`: Converte uma coluna para o tipo datetime, transformando erros em `NaT`.

#### 1.5 Feature Engineering
*   `dados.apply(lambda x: minha_funcao(x['COLUNA1'], x['COLUNA2']),axis=1)`: Aplica uma função a cada linha do DataFrame para criar uma nova coluna.
*   `pd.get_dummies(dados, columns=['COLUNA'])`: Converte colunas categóricas em variáveis dummy (one-hot encoding).
*   `dados.merge(outro_df, on='ID', how='left')`: Combina dois DataFrames com base em uma coluna comum.
*   `dados.rename(columns={'NOME_ANTIGO':'NOME_NOVO'}, inplace=True)`: Renomeia colunas do DataFrame.
*   `dados['COLUNA'].str.contains('texto', case=False)`: Cria uma coluna booleana baseada na presença de um 'texto' (ignorando maiúsculas/minúsculas).

```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```

```python
tabela_cruzada = pd.crosstab(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
tabela_cruzada
```

  * Conexão com o Banco de Dados: Estabelecemos uma conexão com um banco de dados SQLite (status_brasil.db) usando a biblioteca sqlite3.

  * Consultas SQL Avançadas com Agregação e Junção: Construímos consultas mais complexas para calcular a média de renda e educação por estado. Isso envolveu o uso de INNER JOIN para combinar dados de Municipios_Brasileiros e Municipios_Status, GROUP BY para agregar os resultados por estado e funções de agregação como AVG().
  * Parametrização de Consultas: Aprendemos a passar listas de parâmetros para consultas SQL usando o WHERE ... IN com placeholders (?), o que permite filtrar dinamicamente os resultados com base em dados de um DataFrame Pandas.
  * Carregamento de Dados SQL para Pandas: Utilizamos pd.read_sql() para executar as consultas SQL e carregar os resultados diretamente em DataFrames do Pandas, facilitando a manipulação e análise subsequente.
